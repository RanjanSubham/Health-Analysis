```python
health_insurance_count = df[df['Awareness of Health Insurance Plans'] == 'Yes'].shape[0]

# Count the number of individuals aware of retirement plans
retirement_plans_count = df[df['Awareness of Retirement Plans'] != 'No'].shape[0]

# Total number of individuals
total_individuals = df.shape[0]

# Calculate the percentage of individuals aware of health insurance plans
health_insurance_percentage = (health_insurance_count / total_individuals) * 100

# Calculate the percentage of individuals aware of retirement plans
retirement_plans_percentage = (retirement_plans_count / total_individuals) * 100

print("Percentage of individuals aware of health insurance plans: {:.2f}%".format(health_insurance_percentage))
print("Percentage of individuals aware of retirement plans: {:.2f}%".format(retirement_plans_percentage))
```
